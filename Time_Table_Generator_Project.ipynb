## Generating time slots

```python
# Initialize timetable dictionary
grades = ["Jr. KG", "Sr. KG", "I"]
divisions = ["A", "B", "C", "D"]
timetable = {grade: {div: [] for div in divisions} for grade in grades}

# Define time slots for timetable
def generate_time_slots(start_time, end_time, period_duration, lunch_breaks, assembly_time):
    current_time = start_time
    slots = []
    assembly_time = datetime.strptime(assembly_time, "%H:%M") if assembly_time else None
    lunch_breaks = [(datetime.strptime(start, "%H:%M"), datetime.strptime(end, "%H:%M")) for start, end in lunch_breaks]

    while current_time < end_time:
        # Skip lunch breaks
        if any(start <= current_time < end for start, end in lunch_breaks):
            current_time += timedelta(minutes=period_duration)
            continue
        # Skip assembly time
        if assembly_time and current_time.time() == assembly_time.time():
            current_time += timedelta(minutes=period_duration)
            continue
        # Add time slot
        slots.append(current_time.strftime("%H:%M"))
        current_time += timedelta(minutes=period_duration)
    return slots

# Define school timings
school_timings = {
    "Jr. KG": {
        "start": datetime.strptime("10:15", "%H:%M"),
        "end": datetime.strptime("14:15", "%H:%M"),
        "lunch": [("12:45", "13:15")],
        "assembly": "10:45"
    },
    "Sr. KG": {
        "start": datetime.strptime("10:15", "%H:%M"),
        "end": datetime.strptime("14:15", "%H:%M"),
        "lunch": [("12:45", "13:15")],
        "assembly": "10:45"
    },
    "I": {
        "start": datetime.strptime("08:15", "%H:%M"),
        "end": datetime.strptime("14:15", "%H:%M"),
        "lunch": [("09:15", "09:45"), ("12:45", "13:15")],
        "assembly": "10:45"
    }
}

# Generate time slots
period_duration = 30
for grade, timings in school_timings.items():
    time_slots = generate_time_slots(
        timings["start"], timings["end"], period_duration, timings["lunch"], timings["assembly"]
    )
    print(f"\nTime Slots for {grade}: {time_slots}")
```

## Assigning subjects with combinations

```python
def assign_subjects_with_combinations(
    timetable, subject_allotment, teacher_mapping_df1, grade, divisions, slots_per_day, predefined_combinations
):
    # Create a list of required periods per subject for each day of the week
    daily_subject_requirements = {}

    # Process subject allotment to get exact period requirements
    for subject in subject_allotment:
        if subject[grade] and pd.notna(subject[grade]):
            try:
                periods = float(subject[grade])
                if periods > 0:
                    daily_subject_requirements[subject["Subject"]] = periods
            except (ValueError, TypeError):
                continue

    for div in divisions:
        daily_slots = [[] for _ in range(5)]  # 5 days

        # Initialize all slots as None first
        for day in range(5):
            for slot in range(slots_per_day):
                daily_slots[day].append(None)

        # First pass: Assign fixed slots (Assembly and Lunch)
        for day in range(5):
            # Assembly on Day 2, Slot 2
            if day == 1:
                daily_slots[day][1] = ("Assembly", "N/A")

            if grade == "I":
                # Lunch Break for Grade I: Slot 3 and Slot 10
                daily_slots[day][2] = ("Lunch Break", "N/A")
                daily_slots[day][9] = ("Lunch Break", "N/A")
            else:
                # Lunch Break for Jr. KG and Sr. KG: Slot 6
                daily_slots[day][5] = ("Lunch Break", "N/A")

        # Second pass: Assign predefined combinations
        predefined_combination_days = [0, 1, 2]  # Days for predefined combinations (first 3 days)
        predefined_combinations_iter = iter(predefined_combinations)

        # Predefined teacher mappings for Jr. KG and Sr. KG
        predefined_teacher_mapping = {
            "Computer Studies/Circle Time": {
                "Jr. KG A": "Sayantani", "Jr. KG B": "Sayantani", "Jr. KG C": "Sayantani", "Jr. KG D": "Sayantani",
                "Sr. KG A": "Sayantani", "Sr. KG B": "Sayantani", "Sr. KG C": "Sayantani", "Sr. KG D": "Sayantani"
            },
            "Life Skill/Yoga": {
                "Jr. KG A": "Vanita J", "Jr. KG B": "Vanita J", "Jr. KG C": "Vanita J", "Jr. KG D": "Vanita J",
                "Sr. KG A": "Vanita J", "Sr. KG B": "Vanita J", "Sr. KG C": "Vanita J", "Sr. KG D": "Vanita J"
            },
            "Play Pen/Role Play": {
                "Jr. KG A": "Kim, Rukhsaar", "Jr. KG B": "Maria, Shital", "Jr. KG C": "Shireen, Uphaar", "Jr. KG D": "Priyanka, Darshana",
                "Sr. KG A": "Pooja, Nupur", "Sr. KG B": "Anjum Lakshmipriya", "Sr. KG C": "Shubha, Jyoti", "Sr. KG D": "Ankita, Bhakti"
            },
            "DEED/Play Pen": {
                "Sr. KG A": "Ankita", "Sr. KG B": "Ankita", "Sr. KG C": "Ankita", "Sr. KG D": "Ankita"
            }
        }

        # Second pass: Assign predefined combinations and regular subjects
        subject_tracker = {subject: 0 for subject in daily_subject_requirements.keys()}

        for day in range(5):
            for slot in range(slots_per_day):
                if day == 1 and slot == 1:  # Assembly on Day 2, Slot 2
                    daily_slots[day][slot] = ("Assembly", "N/A")
                elif (grade == "I" and (slot == 2 or slot == 9)):
                    # Lunch Break Slot
                    daily_slots[day][slot] = ("Lunch Break", "N/A")
                elif((grade == "Jr. KG" or grade == "Sr. KG")and slot == 5):
                    # Lunch Break Slot
                    daily_slots[day][slot] = ("Lunch Break", "N/A")
                else:
                    # Assign a predefined combination to specific slots (Day 1, Day 2, Day 3)
                    if day in predefined_combination_days:
                        combination = next(predefined_combinations_iter, None)
                        if combination:
                            subject = combination
                            teacher = predefined_teacher_mapping[combination].get(f"{grade} {div}", "TBD")
                            daily_slots[day][slot] = (subject, teacher)
                            predefined_combination_days.remove(day)  # Remove day from available predefined days
                        else:
                            # Fallback to regular subjects if no predefined combinations are left
                            available_subjects = [
                                sub for sub, count in subject_tracker.items() if count < daily_subject_requirements.get(sub, 0)
                            ]
                            if available_subjects:
                                subject = random.choice(available_subjects)
                                teacher_key = f"{grade} {div}"
                                teacher_options = teacher_mapping_df1.loc[
                                    teacher_mapping_df1["Subject"] == subject, teacher_key
                                ]

                                if not teacher_options.empty:
                                    teacher_list = teacher_options.iloc[0]
                                    if isinstance(teacher_list, str):
                                        teacher_list = teacher_list.split(", ")
                                    if teacher_list:
                                        teacher = random.choice(teacher_list)
                                    else:
                                        teacher = "TBD"  # Assign TBD if teacher_list is empty or None
                                else:
                                    teacher = "TBD"
                                daily_slots[day][slot] = (subject, teacher)
                                subject_tracker[subject] += 1
                    else:
                        # Allocate regular subjects
                        available_subjects = [
                            sub for sub, count in subject_tracker.items() if count < daily_subject_requirements.get(sub, 0)
                        ]
                        if available_subjects:
                            subject = random.choice(available_subjects)
                            teacher_key = f"{grade} {div}"
                            teacher_options = teacher_mapping_df1.loc[
                                teacher_mapping_df1["Subject"] == subject, teacher_key
                            ]

                            if not teacher_options.empty:
                                teacher_list = teacher_options.iloc[0]
                                if isinstance(teacher_list, str):
                                    teacher_list = teacher_list.split(", ")
                                if teacher_list:
                                    teacher = random.choice(teacher_list)
                                else:
                                    teacher = "TBD"  # Assign TBD if teacher_list is empty or None
                            else:
                                teacher = "TBD"
                            daily_slots[day][slot] = (subject, teacher)
                            subject_tracker[subject] += 1

        # Fill remaining slots with "Free Period"
        for day in range(5):
            for slot in range(slots_per_day):
                if daily_slots[day][slot] is None:
                    daily_slots[day][slot] = ("Free Period", "N/A")

        timetable[grade][div] = daily_slots
```

## Slots per day

```python
# Initialize timetable
grades = ["Jr. KG", "Sr. KG", "I"]
divisions = ["A", "B", "C", "D"]
timetable = {grade: {div: [] for div in divisions} for grade in grades}

# Define slots per day for each grade
slots_per_day = {
    "Jr. KG": 8,
    "Sr. KG": 8,
    "I": 12
}
```

## Fractional periods are excluded from the data to handle them properly

```python
# Predefined combinations for Jr. KG and Sr. KG
predefined_combinations_JrKG = ["Computer Studies/Circle Time", "Life Skill/Yoga", "Play Pen/Role Play"]
predefined_combinations_SrKG = ["Computer Studies/Circle Time", "Life Skill/Yoga", "DEED/Play Pen"]
```

## Generating Timetable

```python
# Generate timetable for Jr. KG, Sr. KG, and I
for grade in grades:
    predefined_combinations = predefined_combinations_JrKG if grade == "Jr. KG" else predefined_combinations_SrKG if grade == "Sr. KG" else []
    assign_subjects_with_combinations(
        timetable,
        subject_allotment,
        teacher_mapping_df1,
        grade,
        divisions,
        slots_per_day[grade],
        predefined_combinations
    )

# Print and validate the timetable
print("\nGenerated Timetable:")
for grade, divisions in timetable.items():
    print(f"\n{grade}:")
    for div, daily_slots in divisions.items():
        print(f"\nDivision {div}:")
        for day, slots in enumerate(daily_slots, 1):
            print(f"Day {day}: {slots}")
```

## Optional: Validating function ( Used to validate the exact number of periods is assigned to the slots as per subject_allotment data)

```python
# Validation function
def validate_timetable(timetable, subject_allotment):
    for grade in timetable:
        print(f"\nValidating {grade}:")
        for div in timetable[grade]:
            subject_counts = {}
            for day in timetable[grade][div]:
                for subject, _ in day:
                    if subject not in ["Assembly", "Lunch Break", "Free Period"]:
                        subject_counts[subject] = subject_counts.get(subject, 0) + 1

            print(f"\nDivision {div} subject counts:")
            for subject in subject_allotment:
                if subject[grade] and pd.notna(subject[grade]):
                    try:
                        expected = float(subject[grade])
                        if expected > 0:
                            actual = subject_counts.get(subject["Subject"], 0)
                            print(f"{subject['Subject']}: {actual} periods (Expected: {expected})")
                            if actual != int(expected):
                                print(f"WARNING: Mismatch in {subject['Subject']} periods!")
                    except (ValueError, TypeError):
                        continue

# Validation results
print("\nValidation Results:")
validate_timetable(timetable, subject_allotment)
```
